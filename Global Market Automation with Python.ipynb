```python
import requests as req
import os

# quick tool to grab exchange rates and see what's up with the market
class RateChecker:
    def __init__(self, key, start_curr="USD"):
        self.key = key
        # building the link for the get request
        self.link = f"https://v6.exchangerate-api.com/v6/{key}/latest/{start_curr}"
        self.raw_info = None

    def get_rates(self):
        """ grabs the latest json data from the api """
        try:
            res = req.get(self.link)

            # check if the request actually worked
            if res.status_code != 200:
                print(f"Bummer. Status code: {res.status_code}")
                return

            self.raw_info = res.json()
            print(f">>> Got the data for {self.raw_info.get('base_code')}")

        except Exception as err:
            print(f"Connection tanked: {err}")

    def show_stats(self, coins_to_check):
        """ prints out a little table of the rates we care about """
        if not self.raw_info:
            print("Nothing to show yet. Run get_rates first.")
            return

        # digging into the nested dict
        all_rates = self.raw_info.get('conversion_rates', {})

        print("\n--- Market Overview ---")
        print(f"Ticker".ljust(10) + "| Value")
        print("-" * 22)

        for tick in coins_to_check:
            val = all_rates.get(tick)
            if val:
                # formatting to 4 decimals so it looks clean
                print(f"{tick.ljust(10)}| {val:.4f}")
            else:
                print(f"{tick.ljust(10)}| ??? (Not found)")

# --- Main block ---
if __name__ == "__main__":
    # stick your key here
    MY_KEY = "YOUR_API_KEY"

    # Init the checker
    bot = RateChecker(MY_KEY, "USD")

    # Do the work
    bot.get_rates()

    # List of stuff I'm watching right now
    watchlist = ["EUR", "GBP", "JPY", "INR", "BTC"]
    bot.show_stats(watchlist)
```
